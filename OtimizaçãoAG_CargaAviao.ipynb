```
from random import Random
from time import time
from math import cos
from math import pi
from inspyred import ec
from inspyred.ec import terminators
import numpy as np
import os

def generate_(random, args):
    size = args.get('num_inputs', 12)
    return [random.randint(0, 15000) for i in range(size)]

def evaluate_(candidates, args):
    fitness = []
    for cs in candidates:
        fit = perform_fitness(cs[0], cs[1], cs[2], cs[3], cs[4],
                              cs[5], cs[6], cs[7], cs[8], cs[9],
                              cs[10], cs[11])
        fitness.append(fit)
    return fitness

def perform_fitness(C1_D, C1_C, C1_T,
                    C2_D, C2_C, C2_T,
                    C3_D, C3_C, C3_T,
                    C4_D, C4_C, C4_T):

    C1_D = np.round(C1_D)
    C1_C = np.round(C1_C)
    C1_T = np.round(C1_T)

    C2_D = np.round(C2_D)
    C2_C = np.round(C2_C)
    C2_T = np.round(C2_T)

    C3_D = np.round(C3_D)
    C3_C = np.round(C3_C)
    C3_T = np.round(C3_T)

    C4_D = np.round(C4_D)
    C4_C = np.round(C4_C)
    C4_T = np.round(C4_T)

    carga1 = C1_D + C1_C + C1_T
    carga2 = C2_D + C2_C + C2_T
    carga3 = C3_D + C3_C + C3_T
    carga4 = C4_D + C4_C + C4_T

    compartimentoDianteiro = C1_D + C2_D + C3_D + C4_D
    compartimentoCentral = C1_C + C2_C + C3_C + C4_C
    compartimentoTraseiro = C1_T + C2_T + C3_T + C4_T

    # 12.350 -> estimativa de valor superior
    fit = float((0.31*carga1 + 0.38*carga2 + 0.35*carga3 + 0.285*carga4) / 12350)

    # penalizacoes - carga por setor
    h1 = np.maximum(0, float((C1_D + C2_D + C3_D + C4_D) - 10000)) / (10000/10) # compartimento dianteiro -> peso máximo: 10.000
    h2 = np.maximum(0, float((C1_C + C2_C + C3_C + C4_C) - 16000)) / (16000/10) # compartimento central -> peso máximo: 16.000
    h3 = np.maximum(0, float((C1_T + C2_T + C3_T + C4_T) - 8000)) / (8000/10) # compartimento traseiro -> peso máximo: 8.000

    # penalizacoes - peso da carga
    h4 = np.maximum(0, float((C1_D + C1_C + C1_T)-18000)) / (18000/10) # Carga 1 -> peso máximo: 18.000
    h5 = np.maximum(0, float((C2_D + C2_C + C2_T)-15000)) / (15000/10) # Carga 2 -> peso máximo: 15.000
    h6 = np.maximum(0, float((C3_D + C3_C + C3_T)-23000)) / (23000/10) # Carga 3 -> peso máximo: 23.000
    h7 = np.maximum(0, float((C4_D + C4_C + C4_T)-12000)) / (12000/10) # Carga 4 -> peso máximo: 12.000

    # penalizações - volume da carga
    h8 = np.maximum(0, float(C1_D*0.48 + C2_D*0.65
                    + C3_D*0.58 + C4_D*0.39)-6800) / (6800/10) # Compartimento 1 -> volume máximo: 6800m3

    h9 = np.maximum(0, float(C1_C*0.48 + C2_C*0.65
                    + C3_C*0.58 + C4_C*0.39)-8700) / (8700/10) # Compartimento 2 -> volume máximo: 8700m3
                    
    h10 = np.maximum(0, float(C1_T*0.48 + C2_T*0.65
                    + C3_T*0.58 + C4_T*0.39)-5300) / (5300/10) # Compartimento 3 -> volume máximo: 5300m3


    # penalizações - proporção das cargas
    h11 = np.maximum(0, float((compartimentoDianteiro/34000) - (10000/34000))) / float((10000/34000)/13) # dianteir

    h12 = np.maximum(0, float((compartimentoCentral/34000) - (16000/34000))) / float((16000/34000)/13) # central

    h13 = np.maximum(0, float((compartimentoTraseiro/34000) - (8000/34000))) / float((8000/34000)/13) # traseiro

    fit = fit - (h1 + h2 + h3 + h4 + h5 + h6 + h7 + h8 + h9 + h10 + h11 + h12 + h13)
  
    return fit

def solution_evaluation(C1_D, C1_C, C1_T,
                    C2_D, C2_C, C2_T,
                    C3_D, C3_C, C3_T,
                    C4_D, C4_C, C4_T):

    C1_D = np.round(C1_D)
    C1_C = np.round(C1_C)
    C1_T = np.round(C1_T)

    C2_D = np.round(C2_D)
    C2_C = np.round(C2_C)
    C2_T = np.round(C2_T)

    C3_D = np.round(C3_D)
    C3_C = np.round(C3_C)
    C3_T = np.round(C3_T)

    C4_D = np.round(C4_D)
    C4_C = np.round(C4_C)
    C4_T = np.round(C4_T)

    print("########## RESUMO ##########")
    print("Lucro total {}".format(float(0.31*(C1_D + C1_C + C1_T)
                                        + 0.38*(C2_D + C2_C + C2_T)
                                        + 0.35*(C3_D + C3_C + C3_T)
                                        + 0.285*(C4_D + C4_C + C4_T))))
    
    print("Espaco utilizado dos compartimentos (Max. 34000): {}".format(float(C1_D + C1_C + C1_T
                                                                       +C2_D + C2_C + C2_T
                                                                       +C3_D + C3_C + C3_T
                                                                       +C4_D + C4_C + C4_T)))
    print("------------------------------------------------------------")
    print("Carga C1 (Max. 18000) {}".format(C1_D + C1_C + C1_T))
    print("Carga C2 (Max. 15000) {}".format(C2_D + C2_C + C2_T))
    print("Carga C3 (Max. 23000) {}".format(C3_D + C3_C + C3_T))
    print("Carga C4 (Max. 12000) {}".format(C4_D + C4_C + C4_T))
    print("------------------------------------------------------------")
    print("R$ - Carga C1 {}".format((C1_D + C1_C + C1_T) * 0.31))
    print("R$ - Carga C2 {}".format((C2_D + C2_C + C2_T) * 0.38))
    print("R$ - Carga C3 {}".format((C3_D + C3_C + C3_T) * 0.35))
    print("R$ - Carga C4 {}".format((C4_D + C4_C + C4_T) * 0.285))
    print("------------------------------------------------------------")
    print("Carga Dianteira (Max. 10000) {}".format(C1_D + C2_D + C3_D + C4_D))
    print("Carga Cenral (Max. 16000) {}".format(C1_C + C2_C + C3_C + C4_C))
    print("Carga Traseira (Max. 8000) {}".format(C1_T + C2_T + C3_T + C4_T))
    print("------------------------------------------------------------")
    print("Carga 1 - Dianteira: {} | Central: {} | Traseira: {}".format(C1_D, C1_C, C1_T))
    print("Carga 2 - Dianteira: {} | Central: {} | Traseira: {}".format(C2_D, C2_C, C2_T))
    print("Carga 3 - Dianteira: {} | Central: {} | Traseira: {}".format(C3_D, C3_C, C3_T))
    print("Carga 4 - Dianteira: {} | Central: {} | Traseira: {}".format(C4_D, C4_C, C4_T))


def main():
    rand = Random()
    rand.seed(int(time()))

    ea = ec.GA(rand)
    ea.selector = ec.selectors.tournament_selection
    ea.variator = [ec.variators.uniform_crossover,
                   ec.variators.gaussian_mutation]
    ea.replacer = ec.replacers.steady_state_replacement

    ea.terminator = terminators.generation_termination
    ea.observer = [ec.observers.stats_observer, ec.observers.file_observer]

    final_pop = ea.evolve(generator=generate_,
                          evaluator=evaluate_,
                          pop_size=3000,
                          maximize=True,
                          bounder=ec.Bounder(0, 15000),
                          max_generations=100000,
                          num_inputs=12,
                          crossover_rate=1.0,
                          num_crossover_points=1,
                          mutation_rate=0.25,
                          num_elites=3,
                          num_selected=12,
                          tournament_size=12,
                          statistics_file=open('cargas_stats.csv', 'w'),
                          individuals_file=open('cargas_individuals.csv', 'w'))

    final_pop.sort(reverse=True)

    perform_fitness(final_pop[0].candidate[0], final_pop[0].candidate[1], final_pop[0].candidate[2], final_pop[0].candidate[3],
                    final_pop[0].candidate[4], final_pop[0].candidate[5], final_pop[0].candidate[6], final_pop[0].candidate[7],
                    final_pop[0].candidate[8], final_pop[0].candidate[9], final_pop[0].candidate[10], final_pop[0].candidate[11])
    
    solution_evaluation(final_pop[0].candidate[0], final_pop[0].candidate[1], final_pop[0].candidate[2], final_pop[0].candidate[3],
                    final_pop[0].candidate[4], final_pop[0].candidate[5], final_pop[0].candidate[6], final_pop[0].candidate[7],
                    final_pop[0].candidate[8], final_pop[0].candidate[9], final_pop[0].candidate[10], final_pop[0].candidate[11])

main()
```
